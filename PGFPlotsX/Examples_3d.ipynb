```julia
using LaTeXStrings
```

```julia
x = range(-2; stop = 2, length = 20)
y = range(-0.5; stop = 3, length = 25)
f(x, y) = (1 - x)^2 + 100*(y - x^2)^2

@pgf Axis({ height = "10cm", width = "15cm" },
  Plot3({surf,},
    Coordinates(x, y, f.(x, y'))
  )
)
```

```julia
@pgf Axis({ height = "15cm", width = "15cm", view=(30,10) },
  Plot3(
    {
        surf,
        shader = "interp",
    },
    Coordinates(x, y, reshape(Vpot,(30,30)))
  )
)
```

Using LaTeX string

```julia
@pgf Axis(
  {
    height = "15cm",
    width = "15cm",
    view = (30,10),
    title = L"x \exp(-x^2 - y^2)",
    xlabel = L"x",
    ylabel = L"y",
  },
  Plot3( {
    surf,
    domain = "-2:2",
    domain_y = "-1.3:1.3",
  }, Expression("x * exp(-x^2 - y^2)")
  )
)
```
